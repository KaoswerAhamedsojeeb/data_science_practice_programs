```python
import json
```

```python
def load_data(file_name):
    with open(file_name, "r") as f:
        data = json.load(f)
        return data
```

```python
loaded_data = load_data("sample.json")
loaded_data
```

```python
def clean_data (data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    for user in data:
        #turning text rating to number
        rating = str(user["ratings"]).strip().lower()
        if(rating in text_to_num):
            rating = text_to_num[rating]
        user["ratings"] = rating
        #handling missing values(age)
        user_age = user.get("age")
        if(user_age== None):
            user["age"] = None
```

```python
cleaned_data = clean_data(loaded_data)
```
