```python
def find_max(data):
    current_max = data[0]
    for i in data:
        if current_max < i : 
            current_max = i 
    return current_max     
```

```python
print(find_max(df["unit_price"]))
```

```python
fitness_data = pd.read_csv("fitness_data_set.csv")
```

```python
fitness_data.groupby("BMIcase")["BMI"].min()
```

```python
fitness_data.groupby("Gender")["BMI"].agg(["mean","min","max"])
```

```python

fitness_data.groupby("BMIcase")["Weight"].aggregate(average_weight = "mean", minimum_weight= "min", maximum_weight = "max")
```

```python
fitness_data.groupby("BMIcase").agg({
    "Weight":"mean",
    "Age": "max",
    "Height": "min"
})
```

```python

fitness_data.groupby("BMIcase").agg(
    average_weight=("Weight","mean"),
    maximum_age=("Age","max"),
    minimum_height= ("Height","min")
)

```
