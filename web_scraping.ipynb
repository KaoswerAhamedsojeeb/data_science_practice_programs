```python
with open("scraped data/scraped_quote.html", "r", encoding="utf-8") as f:
    contents = f.read()

soup = BeautifulSoup(contents, "lxml")
```

```python
all_p = soup.find_all("p")
for p in all_p:
    print(p.get_text(strip=True))
```

```python
all_quote = soup.find_all('div',class_="quote")

for q in all_quote:
    print(q.get_text(strip =True))
```

```python
quotes = []
author= []

for i in range(1,11):
    URL = "https://quotes.toscrape.com/page/{}/".format(i)
    res = requests.get(URL).text
    soup = BeautifulSoup(res, 'lxml')
    for j in soup.find_all('div',class_="quote"):
         quotes.append(j.find('span',class_="text").text)
         author.append(j.find('small', class_='author').text)
```

```python
df = pd.DataFrame({"author_naem": author, "Quotes": quotes})
```
