```python
res = requests.get(URL)
```

```python
# prefessonal way, handiling error 

try:
    response = requests.get(URL, timeout=5)
    response.raise_for_status() # raises error for 4xx and 5xx
except requests.exceptions.RequestException as e:
    print("Error", e)
    
```
